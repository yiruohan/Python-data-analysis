```python
plt.hist(s1)
```

```python
s1.plot(kind='kde')
```

```python
sns.distplot(s1, hist=True, kde=True)
```

```python
sns.distplot(s1, hist=True, kde=False)
```

```python
sns.distplot(s1, hist=False, kde=True)
```

```python
sns.distplot(s1, hist=False, kde=True, rug=True)
```
